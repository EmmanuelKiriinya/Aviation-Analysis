## Data Cleaning  
In this section we'll be carrying out data cleaning. This includes dropping irrelevant columns and rows etc.

```python
#Make a copy of the dataset and print first 5 entries.

"""This is essential especially when you make a miastake is easier to correct since you're working with 
a copy of the original dataset"""
aviation = df.copy()
aviation.head()
```

```python
# Drop duplicates
aviation.drop_duplicates(inplace=True)
```

```python
# Dataset info
aviation.info()
```

```python
# Converting the Event date to year and extracting the year
aviation["Event.Date"] = pd.to_datetime(aviation["Event.Date"])
```

```python
aviation.head()
```

```python
# Checking number of nulls
aviation.isnull().sum()
```

```python
aviation["Air.carrier"].unique()
```

```python
# Drop irrelevant columns
aviation = aviation.drop(columns = ["Event.Id", "Accident.Number", "Latitude"
        , "Longitude", "Airport.Code", "Airport.Name", "Location", "Registration.Number", "Schedule", "FAR.Description", "Publication.Date", "Air.carrier"])
aviation.head()
```

```python
# Calculate the 5% threshold
threshold = (5 / 100) * len(aviation)

# Loop through columns
for column in aviation.columns:
    num_missing = aviation[column].isnull().sum()
    if num_missing < threshold:
        aviation = aviation.dropna(subset=[column])

        



```

```python
aviation["Aircraft.Category"].unique()
```

```python
# Filling the nulls in the Aircraft Category with unknown
aviation["Aircraft.Category"] = aviation["Aircraft.Category"].fillna(value="Unknown")

# Filling Total.Fatal.Injuries, Total.Serious.Injuries, Total.Minor.Injuries, Total.Uninjured with zeroes assuming they're 0
injury_cols = ["Total.Fatal.Injuries", "Total.Serious.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]

aviation[injury_cols] = aviation[injury_cols].fillna(0)

# Replace phase of flight nulls with the mode
mode1 = aviation["Broad.phase.of.flight"].mode()[0]
mode1
aviation["Broad.phase.of.flight"] = aviation["Broad.phase.of.flight"].fillna(mode1)
```

```python
# Removing incosistencies in the Make column
aviation['Make'] = aviation['Make'].str.lower()
```

```python
# Cnfirming all nulls have been removed
aviation.isnull().sum()
```

```python
cleaned_dataset = aviation.to_csv("cleaned.csv")

```

## EDA-Exploratory Data Analysis

```python
# Plotting Accidents over period of time
aviation['Year'] = aviation['Event.Date'].dt.year # Extracting the year
accidents_per_year = aviation.groupby('Year').size().reset_index(name='Count')

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=accidents_per_year, x='Year', y='Count', marker='o')
plt.title('Trend of Aviation Accidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting injury severity distribution
print(aviation['Injury.Severity'].unique()) # Checking the injury distribution

aviation['Injury.Severity'] = aviation['Injury.Severity'].str.extract(r'(\w+)', expand=False)
```

```python
print(aviation['Injury.Severity'].unique())
```

```python
# Top Aircraft manufacturers involved
aviation["Make"].unique()
# Plot
top_makes = aviation['Make'].value_counts().head(10)

#  Plot as horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_makes.values, y=top_makes.index, palette='tab10')
plt.title('Top 10 Aircraft Manufacturers Involved in Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Aircraft Manufacturer')
plt.tight_layout()
plt.show()
```

```python
# Accidents by phase of flight
aviation["Broad.phase.of.flight"].unique() # Checking the distinct entries

# Plot
phase_counts = df['Broad.phase.of.flight'].value_counts()

# Bar plot of accident counts per phase of flight
plt.figure(figsize=(10, 5))
sns.barplot(x=phase_counts.index, y=phase_counts.values, palette='Paired')
plt.title('Accidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 countries with most flights
top_countries = aviation['Country'].value_counts().head()

#  Plot as horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='tab10')
plt.title('Top 10 countries with most accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('country')
plt.tight_layout()
plt.show()
```

```python
# Air accidents per purpose of flight
aviation["Purpose.of.flight"].unique() # Checking unique entries
purpose_of_flight = aviation["Purpose.of.flight"].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=purpose_of_flight.values, y=purpose_of_flight.index, palette='tab10')
plt.title('Air accidents per purpose of flight')
plt.xlabel('Number of Accidents')
plt.ylabel('Purpose of flight')
plt.tight_layout()
plt.show()
```

```python
# Accidents by Amateur planes (Homebuilt planes)
aviation["Amateur.Built"].unique() # Checkig unique entries
amateur = aviation["Amateur.Built"].value_counts()

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(amateur, labels=amateur.index, autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62'])
plt.title('Distribution of accidents by Amateur make')
plt.axis('equal')
plt.show()
```

```python
# Accidents by engine type
aviation["Engine.Type"].unique() # Checking distinct figures

engine_type = aviation['Engine.Type'].value_counts().head()

#plot
plt.figure(figsize=(10, 6))
sns.barplot(x=engine_type.values, y=engine_type.index, palette='tab10')
plt.title('Air accidents by engine type')
plt.xlabel('Number of Accidents')
plt.ylabel('Engine type')
plt.tight_layout()
plt.show()
```

## Insights and Conclusions

- **Aircraft incidents by manufacturer**. Most of the recorderded aircraft accidents are by the manufacturer cessna closely followed by the piper.

- **Accidents by phase of flight**. It is recorded that most air accidents happen during the landing phase of flight closely followed by takeoff phase of flight.

- **Country with the most recorded air accidents**. The United states has the most recorded flight accidents between 1948-203.

- **Air accidents by purpose of flight**. Personal flights have the most recorded fight incidents with about 58000 incidents followed by instructional flights wth the recorded incidents being around 9000.

- **Accidents by engine type**. Reciprocating engines have more accident cases compared to other engine types.


- **Air accidents by amateur make**. Amateur planes or Home-made planes contribute to about 10.5% of the total recorded flight incidents.
